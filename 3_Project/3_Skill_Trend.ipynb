Cómo está la demanda de habilidades para analistas de datos?

```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_US = df[(df['job_country']=='United States') & (df['job_title_short']=='Data Analyst')].copy()
```

```python
df_DA_US['job_posted_month_no'] = df_DA_US['job_posted_date'].dt.month
```

```python
df_DA_US_explode = df_DA_US.explode('job_skills')
```

```python
df_DA_US_pivot = df_DA_US_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)

df_DA_US_pivot.loc['Total'] = df_DA_US_pivot.sum()

df_DA_US_pivot = df_DA_US_pivot[df_DA_US_pivot.loc['Total'].sort_values(ascending=False).index]

df_DA_US_pivot = df_DA_US_pivot.drop('Total')

df_DA_US_pivot
```

```python
df_totals = df_DA_US.groupby('job_posted_month_no').size()
```

```python
df_DA_US_perc = df_DA_US_pivot.div(df_totals/100, axis=0)
```

```python
df_DA_US_perc = df_DA_US_perc.reset_index()
df_DA_US_perc['job_posted_month_no'] = df_DA_US_perc['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_US_perc = df_DA_US_perc.set_index('job_posted_month_no')
df_DA_US_perc = df_DA_US_perc.drop(columns='job_posted_month_no')

df_DA_US_perc
```

```python
df_DA_US_perc
```

```python
df_plot = df_DA_US_perc.iloc[:, :5]

sns.lineplot(data=df_plot, dashes=False, palette='tab10')
sns.set_theme(style='ticks')
sns.despine()

plt.title('Mejores Habilidades para un Analista de Datos')
plt.ylabel('Probabilidad de que lo pida un puesto')
plt.xlabel('2023')
plt.legend().remove()

from matplotlib.ticker import PercentFormatter
ax = plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))

from adjustText import adjust_text

texts = []
for i, col in enumerate(df_plot.columns):
    texts.append(plt.text(11.2, df_plot.iloc[-1, i], col))
adjust_text(texts, arrowprops=dict(arrowstyle="-", color='gray'))

plt.show()
```
